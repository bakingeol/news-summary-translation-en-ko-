```
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer, Adafactor
from torch.utils.data import DataLoader, Dataset, random_split
from tqdm import tqdm
import json

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# 입력과 정답을 처리하는 함수를 정의합니다.
class NewsSummaryDataset(Dataset):
    def __init__(self, data_file, tokenizer):
        with open(data_file, 'r') as f:
            data = json.load(f)

        self.input_ids = data['input_ids']
        self.attention_mask = data['attention_mask']
        self.labels = data['labels']
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        input_ids = torch.tensor(self.input_ids[idx])
        attention_mask = torch.tensor(self.attention_mask[idx])
        label = torch.tensor(self.labels[idx])
        return input_ids, attention_mask, label

# 입력과 정답을 분리합니다.
train_data = data[:1800]
val_data = data[1800:]

# T5Tokenizer 불러오기
tokenizer = AutoTokenizer.from_pretrained('google/flan-t5-base')

# 학습 데이터와 검증 데이터를 T5 모델이 입력받을 수 있는 형식으로 변환
train_encodings = tokenizer([data['text'] for data in train_data], truncation=True, padding=True)
train_labels = tokenizer([data['summary'] for data in train_data], truncation=True, padding=True)
val_encodings = tokenizer([data['text'] for data in val_data], truncation=True, padding=True)
val_labels = tokenizer([data['summary'] for data in val_data], truncation=True, padding=True)

# 데이터를 JSON 파일로 저장
with open(train_params['train_file'], 'w') as f:
    json.dump({'input_ids': train_encodings['input_ids'], 'attention_mask': train_encodings['attention_mask'], 'labels': train_labels['input_ids']}, f)

with open(train_params['val_file'], 'w') as f:
    json.dump({'input_ids': val_encodings['input_ids'], 'attention_mask': val_encodings['attention_mask'], 'labels': val_labels['input_ids']}, f)

# 데이터셋을 생성합니다.
train_dataset = NewsSummaryDataset(train_params['train_file'], tokenizer)
val_dataset = NewsSummaryDataset(train_params['val_file'], tokenizer)

# 데이터로더를 생성합니다.
batch_size = 4
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# 모델을 불러옵니다.
model = T5ForConditionalGeneration.from_pretrained('google/flan-t5-base', cache_dir='path/to/cache/dir')
```

```
from transformers import get_linear_schedule_with_warmup
from transformers import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=len(train_dataloader) * 5)
```

```
# 에폭 수를 정의합니다.
num_epochs = 5

# 각 에폭마다 train_loss와 val_loss를 기록할 리스트를 생성합니다.
train_losses = []
val_losses = []
torch.cuda.empty_cache()
import os

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = '0:128'

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# tqdm 라이브러리를 이용하여 학습 진행 상황을 시각화합니다.
for epoch in tqdm(range(num_epochs), desc="Epoch"):

    model.train()
    train_loss = 0

    for batch in tqdm(train_dataloader, desc="Training", leave=False):
        batch = tuple(b.to(device) for b in batch)
        inputs = {"input_ids": batch[0], "attention_mask": batch[1], "labels": batch[2]}
        outputs = model(**inputs)
        loss = outputs[0]
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        train_loss += loss.item()

    train_loss = train_loss / len(train_dataloader)
    train_losses.append(train_loss)

    model.eval()
    val_loss = 0

    for batch in tqdm(val_dataloader, desc="Validation", leave=False):
        batch = tuple(b.to(device) for b in batch)
        inputs = {"input_ids": batch[0], "attention_mask": batch[1], "labels": batch[2]}
        with torch.no_grad():
            outputs = model(**inputs)
        loss = outputs[0]
        val_loss += loss.item()

    val_loss = val_loss / len(val_dataloader)
    val_losses.append(val_loss)

    print(f"Epoch {epoch+1}")
    print(f"Train loss: {train_loss:.3f}")
    print(f"Validation loss: {val_loss:.3f}")
```
